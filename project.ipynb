```python
#Seperating the data and labels
X = df.drop(columns = 'Personality', axis = 1)
Y = df['Personality']
```

```python
#Data standardisation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
#Plotting this in descending order for better understanding of this visualization

cnt_srs = df['Personality'].value_counts() 
plt.figure(figsize=(12,4)) 
sns.barplot(cnt_srs.index, cnt_srs.values, alpha=0.8) 
plt.xlabel('Personality types', fontsize=12)
plt.ylabel('Numbers', fontsize=12) 
plt.show()
```

# SPLIT DATA IN TRAIN AND TEST

```python
#Train,Test,Split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, stratify = Y,random_state = 2)
```

# MODEL EVALUATION WITH SVM(SUPPORT VECTOR MACHINE)

```python
#Model Training
from sklearn import svm
model = svm.SVC(kernel = 'linear')
```

```python
model.fit(X_train, Y_train)
```

```python
#Finding the accuracy score on training dataset
from sklearn.metrics import accuracy_score
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
train_data_accuracy
```

```python
#Finding the accuracy score on test dataset
from sklearn.metrics import accuracy_score
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
test_data_accuracy
```
